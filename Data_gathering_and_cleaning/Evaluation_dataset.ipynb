```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Sentence pairs in Croatian-Dutch - 2024-03-04.tsv', sep='\t', names=['id_hr', 'hr','id_nl', 'nl'], header=None)
```

```python
df.drop(columns=['id_hr','id_nl'],inplace=True)
```

```python
df.rename(columns={'hr':'Croatian', 'nl':'Dutch'}, inplace=True)
```

```python
# File paths
dutch_dev_path = '/content/flores101_dataset/dev/nld.dev'
croatian_dev_path = '/content/flores101_dataset/dev/hrv.dev'
dutch_devtest_path = '/content/flores101_dataset/devtest/nld.devtest'
croatian_devtest_path = '/content/flores101_dataset/devtest/hrv.devtest'

# Function to read files into DataFrame
def read_file_to_df(file_path, column_name):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.read().splitlines()
    return pd.DataFrame(lines, columns=[column_name])

# Creating DataFrames
dutch_dev_df = read_file_to_df(dutch_dev_path, 'Dutch')
croatian_dev_df = read_file_to_df(croatian_dev_path, 'Croatian')
dutch_devtest_df = read_file_to_df(dutch_devtest_path, 'Dutch')
croatian_devtest_df = read_file_to_df(croatian_devtest_path, 'Croatian')

# Combining DataFrames
dutch_combined_df = pd.concat([dutch_dev_df, dutch_devtest_df], ignore_index=True)
croatian_combined_df = pd.concat([croatian_dev_df, croatian_devtest_df], ignore_index=True)
```

```python
aligned_df = pd.concat([dutch_combined_df, croatian_combined_df], axis=1)
```

```python
aligned_df = aligned_df.append(df)
```
