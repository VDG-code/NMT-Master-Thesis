```python
import re

def split_text_by_punctuation(text):
    # Check if the input is a string
    if not isinstance(text, str):
        return text
    # This regex replaces each punctuation character with the same character surrounded by spaces
    text_with_spaced_punctuation = re.sub(r'([^\w\s])', r' \1 ', text)
    # This line removes extra spaces around the inserted ones, including potential double spaces
    text_normalized_spaces = re.sub(r'\s+', ' ', text_with_spaced_punctuation).strip()
    return text_normalized_spaces
```

```python
merged_df.hr = merged_df.hr.apply(split_text_by_punctuation)
```

```python
merged_df.nl = merged_df.nl.apply(split_text_by_punctuation)
```

```python
merged_df.hr = merged_df.hr.astype(str)
```

```python
merged_df.nl = merged_df.nl.astype(str)
```

```python
# Normalization function for Dutch
def normalize_dutch(text):
    pattern = r'[^a-zA-Z0-9 \n\.,ëËïÏäÄöÖüÜáÁéÉíÍóÓúÚ]'
    return re.sub(pattern, '', text)

# Normalization function for Croatian
def normalize_croatian(text):
    pattern = r'[^a-zA-Z0-9 \n\.,čČćĆđĐšŠžŽ]'
    return re.sub(pattern, '', text)

# Apply normalization to each column
merged_df['nl'] = merged_df['nl'].apply(normalize_dutch)
merged_df['hr'] = merged_df['hr'].apply(normalize_croatian)
```

```python
# Iterate over each row in the DataFrame column
def unique_chars(df, language):
    # Initialize an empty set to store unique characters
    unique_chars_set = set()
    for item in df[language]:
        # Update the set with characters from the current row
        if isinstance(item, str):
            unique_chars_set.update(set(item))

    # Convert the set to a sorted list to have a consistent order
    sorted_unique_chars = sorted(list(unique_chars_set))
    return sorted_unique_chars
```

```python
from sentence_transformers import SentenceTransformer, util
import torch
from tqdm.auto import tqdm

# Ensure CUDA is available and set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Enable automatic mixed precision
torch.cuda.amp.autocast(enabled=True)

# Load the model and move it to the GPU
model = SentenceTransformer('LaBSE').to(device)

# Function to compute cosine similarity scores in batches
def compute_cosine_scores(sentences1, sentences2):
    with torch.no_grad():  # Disable gradient calculation for inference
        embeddings1 = model.encode(sentences1, convert_to_tensor=True, batch_size=256).to(device)
        embeddings2 = model.encode(sentences2, convert_to_tensor=True, batch_size=256).to(device)
        cosine_scores = util.pytorch_cos_sim(embeddings1, embeddings2)
    return cosine_scores

# Batch processing
batch_size = 8192  # Adjust based on GPU memory
alignment_scores = []

for start_idx in tqdm(range(0, len(merged_df), batch_size)):
    end_idx = min(start_idx + batch_size, len(merged_df))
    batch_hr = merged_df['hr'][start_idx:end_idx].tolist()
    batch_nl = merged_df['nl'][start_idx:end_idx].tolist()

    # Compute scores with mixed precision context
    with torch.cuda.amp.autocast():
        scores = compute_cosine_scores(batch_hr, batch_nl)

    alignment_scores.extend(scores.diag().cpu().tolist())

# Assign scores back to DataFrame and filter
merged_df['alignment_score'] = alignment_scores
highly_aligned_df = merged_df[merged_df['alignment_score'] > 0.8]

print(highly_aligned_df)
```

```python
# Function to replace multiple whitespace with single whitespace
def normalize_whitespace(text):
    # Replace one or more whitespace (spaces, tabs, newlines) with a single space
    return re.sub(r'\s+', ' ', text)
```

```python
# Apply the normalization function to the 'text' column
highly_aligned_df['nl'] = highly_aligned_df['nl'].apply(normalize_whitespace)
highly_aligned_df['hr'] = highly_aligned_df['hr'].apply(normalize_whitespace)
```
