```python
from datasets import load_dataset
```

```python
bible_dataset = load_dataset("bible_para", lang1="hr", lang2="nl")
```

```python
bible_dataset
```

```python
import pandas as pd

train_data = bible_dataset['train']

# Convert the 'train' split data into a DataFrame
df = pd.DataFrame(train_data)

# Extract and expand the 'translation' column into 'hr' and 'nl' columns
df['hr'] = df['translation'].apply(lambda x: x.get('hr', ''))
df['nl'] = df['translation'].apply(lambda x: x.get('nl', ''))

# Drop the original 'translation' column
df = df.drop('translation', axis=1)

# Display the first row or any specific row of your choice
df.iloc[0]
```

```python
multipara_dataset = load_dataset("multi_para_crawl", lang1="hr", lang2="nl")
```

```python
para_data = multipara_dataset['train']

# Convert the 'train' split data into a DataFrame
para_df = pd.DataFrame(para_data)

# Extract and expand the 'translation' column into 'hr' and 'nl' columns
para_df['hr'] = para_df['translation'].apply(lambda x: x.get('hr', ''))
para_df['nl'] = para_df['translation'].apply(lambda x: x.get('nl', ''))

# Drop the original 'translation' column
para_df = para_df.drop('translation', axis=1)

# Display the first row or any specific row of your choice
para_df.iloc[0]
```

```python
ted_talks_2014 = load_dataset("ted_talks_iwslt", language_pair=("hr", "nl"), year="2014")
```

```python
ted_talks_2015 = load_dataset("ted_talks_iwslt", language_pair=("hr", "nl"), year="2015")
```

```python
ted_talks_2016 = load_dataset("ted_talks_iwslt", language_pair=("hr", "nl"), year="2016")
```

```python
def create_translation_df(dataset):
    # Convert the 'train' split data into a DataFrame
    df = pd.DataFrame(dataset['train'])

    # Extract and expand the 'translation' column into 'hr' and 'nl' columns
    df['hr'] = df['translation'].apply(lambda x: x.get('hr', ''))
    df['nl'] = df['translation'].apply(lambda x: x.get('nl', ''))

    # Drop the original 'translation' column
    df = df.drop('translation', axis=1)

    return df
```

```python
# Convert datasets to DataFrames
df_ted_talks_2014 = create_translation_df(ted_talks_2014)
df_ted_talks_2015 = create_translation_df(ted_talks_2015)
df_ted_talks_2016 = create_translation_df(ted_talks_2016)

# Display the first row of each DataFrame
print(df_ted_talks_2014.iloc[0])
print(df_ted_talks_2015.iloc[0])
print(df_ted_talks_2016.iloc[0])
```

```python
# Concatenate all DataFrames
merged_df = pd.concat([df, para_df, df_ted_talks_2014, df_ted_talks_2015, df_ted_talks_2016], axis=0)

# Reset the index of the merged DataFrame
merged_df.reset_index(drop=True, inplace=True)

# Display the first few rows of the merged DataFrame
print(merged_df.head())
```
